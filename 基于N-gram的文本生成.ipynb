```python
# 加载数据集
from datasets import load_from_disk
data_path = "./data/data"
poetry = load_from_disk(data_path)['poetry']
poetry
```

```python
# 开始进行分词行动
import jieba
dataset = []
for sentence in poetry:
    result = jieba.lcut(sentence, cut_all=False)
    dataset.append(result)
dataset[0]
```

```python
import random
from collections import defaultdict

class NGramTextGenerator:
    def __init__(self, n=2):
        self.n = n
        self.ngram_freq = defaultdict(list)

    def train(self, dataset):
        for words in dataset:
            for i in range(len(words) - self.n):
                prefix = tuple(words[i:i+self.n-1])
                next_word = words[i+self.n-1]
                self.ngram_freq[prefix].append(next_word)

    def generate(self, start, length=48):
        words = start
        for _ in range(length):
            prefix = tuple(words[-(self.n-1):])
            if prefix in self.ngram_freq:
                next_word = random.choice(self.ngram_freq[prefix])
                words.append(next_word)
            else:
                break
        return ''.join(words)

# 训练
ngram_gen = NGramTextGenerator(n=3)
ngram_gen.train(dataset)
```

```python
# 开始生成（由于n=3，所以输入的开头句，至少要能被分成3个词）
text = input("诗词开头句: ")
while text != '退出':
    result = jieba.lcut(text, cut_all=False)
    print(f"分词结果:{result}")
    print("结果: " + ngram_gen.generate(result, length=32))
    text = input("诗词开头句: ")
```
